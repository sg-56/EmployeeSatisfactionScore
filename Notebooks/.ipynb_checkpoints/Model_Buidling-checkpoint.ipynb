```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATASET_PATH = '../Data/Employee_data.csv'
df = pd.read_csv(DATASET_PATH)
df.head()
#Getting first 5 rows of the dataset
```

```python
features_not_required = [
    'Employee_ID',
    'Hire_Date',
    'Resigned'
]
```

```python
df.drop(columns=features_not_required,inplace=True)
```

```python
df.info()
```

```python
TARGET_COLUMN = 'Employee_Satisfaction_Score'
```

```python
X = df.drop(columns=[TARGET_COLUMN])
y = df[TARGET_COLUMN]
```

```python
cat_cols = X.select_dtypes(include='object').columns
num_cols = X.select_dtypes(include='number').columns
print(f"Categorical Columns : {cat_cols} \n")
print(f"Numerical Columns : {num_cols}")
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
df[num_cols].describe().T[['min','mean','max']]
```

```python

from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer(sparse=True)
dv.fit(X_train.to_dict(orient='records'))
```

```python
dv.feature_names_
```

```python
X_train_tr = dv.transform(X_train.to_dict(orient='records'))
```

```python
X_test_tr = dv.transform(X_test.to_dict(orient='records'))
```

```python
from xgboost import XGBRegressor
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train_tr,y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
train_preds = model.predict(X_train_tr)
test_preds = model.predict(X_test_tr)
```

```python
r2_score(y_train,train_preds)
```

```python
new_model = XGBRegressor(
    n_esimators = 300,
    max_depth=10,
    n_jobs = -1,
    eval_metric=['rmse']
)
```

```python
new_model.fit(X_train_tr,y_train,verbose=True)
```

```python
train_preds = new_model.predict(X_train_tr)
test_preds = new_model.predict(X_test_tr)
r2_score(y_train,train_preds)
```

```python
import pickle
```

```python
model_path = '../artifacts/model.bin'
preprocessor = '../artifacts/vectorizer.bin'

with open(model_path,'wb') as file:
    pickle.dump(new_model,file)

with open(preprocessor,'wb') as file:
    pickle.dump(dv,file)
```
